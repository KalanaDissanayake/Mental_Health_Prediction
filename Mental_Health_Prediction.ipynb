```python
from sklearn.preprocessing import OrdinalEncoder

# Initialize encoder with explicit order
enc = OrdinalEncoder(categories=[['low', 'medium', 'high']])

# Encode and create new column
data_frame['mental_health_risk_encoded'] = enc.fit_transform(
    data_frame[['mental_health_risk']].apply(lambda s: s.str.strip().str.lower())
).astype(int)

# Remove the original column
data_frame.drop('mental_health_risk', axis=1, inplace=True)
#low=0, medium=1, high=2
```

```python
from sklearn.preprocessing import OrdinalEncoder

# Initialize encoder with correct order so yes=1, no=0
enc = OrdinalEncoder(categories=[['no', 'yes']])

# Encode and create new column
data_frame['mental_health_history_encoded'] = enc.fit_transform(
    data_frame[['mental_health_history']].apply(lambda s: s.str.strip().str.lower())
).astype(int)

# Remove the original column
data_frame.drop('mental_health_history', axis=1, inplace=True)
#yes =1, no=0
```

```python
from sklearn.preprocessing import OrdinalEncoder

# Encode the 'employment_status' column (Unemployed=0,Student=1, Self-employed=2, Employed=3  )
enc = OrdinalEncoder(categories=[['Unemployed', 'Student', 'Self-employed', 'Employed']])

data_frame['employment_status_encoded'] = enc.fit_transform(
    data_frame[['employment_status']].apply(lambda s: s.str.strip())
).astype(int)
data_frame.drop('employment_status', axis=1, inplace=True)
#Unemployed=0,Student=1, Self-employed=2, Employed=3

# Show all columns
data_frame.head()
```

```python
#finding the outliers
def find_outliers_IQR(data_frame):
  q1=data_frame.quantile(0.25)
  q3=data_frame.quantile(0.75)
  IQR=q3-q1
  outliers = data_frame[((data_frame<(q1-1.5*IQR))|(data_frame>(q3+1.5*IQR)))]
  return outliers
```

```python
outliers = find_outliers_IQR(data_frame['stress_level'])
print("number of outliers: "+ str(len(outliers)))
outliers
```

```python
outliers = find_outliers_IQR(data_frame['sleep_hours'])
print("number of outliers: "+ str(len(outliers)))
outliers
```

```python
outliers = find_outliers_IQR(data_frame['physical_activity_days'])
print("number of outliers: "+ str(len(outliers)))
outliers
```

```python
outliers = find_outliers_IQR(data_frame['social_support_score'])
print("number of outliers: "+ str(len(outliers)))
outliers
```
